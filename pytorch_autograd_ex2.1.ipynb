Compute the derivative of sin(ùë•), using Autograd


```python
import numpy as np
import torch
import matplotlib.pyplot as plt

# For x range 0 to 4pi, with increments of 0.1  (hint ‚Äì torch.arange)
x = torch.arange(start=0, end=4*np.pi, step=0.1, requires_grad=True)  # no need for list comprehension, we want one tensor with all values, not a single value tensor
y = torch.sin(x).sum()
deriv = y.backward()  # backward() should be called on a scalar tensor - either give gradient or sum , both make it calculate, then we want to see x.grad()

# %% Plot your result (using matplotlib)
plt.figure()
plt.grid(axis='both')
plt.plot(x.detach().numpy(), x.grad.numpy(), 'b-', label="torch")

# %% Compare to the analytic gradient
a = np.arange(4*np.pi, step=0.1)
plt.plot(a, np.gradient(np.sin(a), 0.1), 'r-', label="analytic")
plt.show()
```
